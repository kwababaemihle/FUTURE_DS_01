```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Sample - Superstore.csv", encoding='latin1')


df.head()
```

```python
df['Month'] = df['Order Date'].dt.strftime('%B')
df['Year'] = df['Order Date'].dt.year
df
```

```python
df = df.dropna(subset=['Order Date', 'Sales', 'Category'])
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df
```

```python
df = df.dropna(subset=['Order Date'])
df
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'])
monthly_sales
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values, marker='o')
plt.title('Monthly Revenue Trend')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title('Sales by Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(7,5))
sns.barplot(x=region_sales.index, y=region_sales.values)
plt.title('Sales by Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_products.index, x=top_products.values)
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.show()
```

```python

df.to_csv('Cleaned_Superstore.csv', index=False)
```

Insights

-Technology category generates the highest revenue.

-The West region contributes the most sales.

-Sales peak in November-December(seasonal trend).

-A few top products generate most revenue.

Recommendations

Focus marketing on top categories.

Increase distribution in high-performing regions.

Run promotions during high-sales months.
